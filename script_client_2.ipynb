```python
# Imports des bibliothèques et modules nécessaires au bon fonctionnement du script
import argparse
import sys # ligne ajoutée pour Google Colab
import joblib
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
import pickle

# Vérifications des types de chaque argument
def checkArguments():
    parser = argparse.ArgumentParser()
    parser.add_argument('-m', '--model', type=str, required=True, help='model') # modèle
    parser.add_argument('-h_tot', '--height', type=float, required=True, help='height') # haut_tot
    parser.add_argument('-trunc_h', '--trunc_height', type=float, required=True, help='trunc_height') #haut_tronc
    parser.add_argument('-trunc_d', '--trunc_diameter', type=float, required=True, help='trunc_diameter') # tronc_diam
    parser.add_argument('-rem', '--remarkable', type=str, required=True, help='remarkable') # remarquable
    parser.add_argument('-state', '--id_state', required=True, help='id_state') # clc_quartier
    parser.add_argument('-secteur', '--id_secteur', type=str, required=True, help='id_secteur') # clc_secteur
    parser.add_argument('-stage', '--id_stage', type=str, required=True, help='id_stage') # fk_stadedev
    parser.add_argument('-etat', '--id_etat', type=str, required=True, help='id_etat') # fk_arb_etat
    parser.add_argument('-feuille', '--id_feuille', type=str, required=True, help='id_feuille') # feuillage
    parser.add_argument('-prec_estim', '--id_prec_estim', type=float, required=True, help='id_prec_estim') # fk_prec_estim
    return parser.parse_args()


# Chargement du modèle demandé
def chargement_modele(model_name):
    model_path = f"gdrive/MyDrive/PROJET_IA/Besoin client 2/{model_name}_model.pkl"
    return joblib.load(model_path)

# Préparation des données pour l'utiliser dans le modèle
def preparer_donnees(args):
    input_data = pd.DataFrame([{
        'clc_quartier': args.id_state,
        'clc_secteur': args.id_secteur,
        'haut_tot': args.height,
        'haut_tronc': args.trunc_height,
        'tronc_diam': args.trunc_diameter,
        'fk_arb_etat': args.id_etat,
        'fk_stadedev': args.id_stage,
        'fk_prec_estim': args.id_prec_estim,
        'feuillage': args.id_feuille,
        'remarquable': args.remarkable,
    }])
    # Colonnes catégorielles à encoder
    colonnes = ['clc_quartier', 'clc_secteur', 'fk_arb_etat', 'fk_stadedev',  'feuillage',
            'remarquable']
    # Encodage des données catégorielles
    label_encoders = {}
    for column in colonnes:
        label_encoders[column] = joblib.load(f'gdrive/MyDrive/PROJET_IA/Besoin client 2/label_encoder_{column}.pkl')
        #print(label_encoders)
        #print(label_encoders[column].classes_)
        input_data[column] = label_encoders[column].fit_transform(input_data[column])
    return input_data

sys.argv=['script_name','--model','random forest regressor',
          '-h_tot','8','-trunc_h','3','-trunc_d','103',
          '-rem','Non',
          '-state',"Quartier de l'Europe",'-secteur','Rue Laplace',
          '-stage','Adulte','-etat','EN PLACE',
          '-feuille','Feuillu',
          '-prec_estim','10'] # ligne ajoutée pour Google Colab

# Main program.
args = checkArguments()
input_data = preparer_donnees(args)

if args.model == 'regression lineaire' or args.model=='régression linéaire':
    model = chargement_modele('regression lineaire')
    prediction = model.predict(input_data)
    print(prediction)
elif args.model == 'decision tree regressor' or args.model=='Decision Tree Regressor' or args.model=="Arbre de décision":
    model = chargement_modele('decision tree regressor')
    prediction = model.predict(input_data)
    print(prediction)
elif args.model == 'random forest regressor' or args.model=='Random Forest Regressor':
    model = chargement_modele('random forest regressor')
    prediction = model.predict(input_data)
    print(prediction)
else:
    print("Modèle non reconnu")
    sys.exit(1)
```
