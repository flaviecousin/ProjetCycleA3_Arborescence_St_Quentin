# **Répartition des données**

```python
# Ouverture et lecture du nouveau dataframe
data2=pd.read_csv("gdrive/MyDrive/PROJET_IA/Besoin client 2/Besoin2.csv")
```

```python
# Séparation des features et de la variable cible
X_n = data2.drop("age_estim",axis=1) # features
Y= data2['age_estim'].copy() # target

# Séparation en base d'apprentissage et de test
X_train, X_test, Y_train, Y_test = train_test_split(X_n, Y, test_size=0.4, train_size = 0.6, random_state=42)
```

# **Modèle régression linéaire**







```python
# Instanciation de la régression linéaire
model_regression_lineaire=LinearRegression()

# Ajustement de la régression linéaire sur la base d'apprentissage
model_regression_lineaire.fit(X_train,Y_train)

# Prédiction des valeurs d'apprentissage
predict = model_regression_lineaire.predict(X_test)

# Enregistrement du modèle
joblib.dump(model_regression_lineaire, 'gdrive/MyDrive/PROJET_IA/Besoin client 2/regression lineaire_model.pkl')

# Calcul de l'erreur quadratique moyenne (RMSE)
print("RMSE : ", sqrt(mean_squared_error(Y,model_regression_lineaire.predict(X_n))))
```

# **Evaluation du modèle d'apprentissage de la régression linéaire**

```python
# Prédiction des valeurs d'apprentissage
y_predict_test=model_regression_lineaire.predict(X_test)

# Calcul du R² pour avoir la précision du modèle
pourcent = r2_score(Y_test, y_predict_test)
print(pourcent)
```
